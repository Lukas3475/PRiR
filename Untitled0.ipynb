## **Metoda prostokątów:**

```
import tensorflow as tf
assert tf.__version__.startswith('2')

def function(x):
    return x * x
intervalBegin = float(input("Podaj dolną granice: "))
intervalEnd = float(input("Podaj górną granice: "))
iterations = int(input("Podaj liczbe iteracji: "))
step = (intervalEnd - intervalBegin) / iterations
integral = 0
for i in range(iterations):
    integral += step * function(intervalBegin + (i - 1) * step)
print("Wynik: ", integral)
```

## **Metoda Simpsona**

```
import tensorflow as tf
assert tf.__version__.startswith('2')
def f(x):
    return 1/(1 + x**2)
def simpson13(x0,xn,n):
    h = (xn - x0) / n
    integration = f(x0) + f(xn)
    for i in range(1,n):
        k = x0 + i*h       
        if i%2 == 0:
            integration = integration + 2 * f(k)
        else:
            integration = integration + 4 * f(k)
    integration = integration * h/3
    return integration
lower_limit = float(input("Podaj dolną granice: "))
upper_limit = float(input("Podaj górną granice: "))
sub_interval = int(input("Podaj liczbe iteracji: "))
result = simpson13(lower_limit, upper_limit, sub_interval)
print("Wynik: " , result)
```

# **Metoda trapezów**

```
import tensorflow as tf
assert tf.__version__.startswith('2')
def f(x):
    return 1/(1 + x**2)
def trapezoidal(x0,xn,n):
    h = (xn - x0) / n
    integration = f(x0) + f(xn)
    for i in range(1,n):
        k = x0 + i*h
        integration = integration + 2 * f(k)
    integration = integration * h/2
    return integration
    
lower_limit = float(input("Podaj dolną granice: "))
upper_limit = float(input("Podaj górną granice: "))
sub_interval = int(input("Podaj liczbe iteracji: "))

result = trapezoidal(lower_limit, upper_limit, sub_interval)
print("Wynik: " , result)
```

```
import tensorflow as tf
import numpy as np
from tensorflow import keras

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([0.0, 8.0, 6.0, 0.0, -4.0, 0.0], dtype=float)
model.fit(xs, ys, epochs=500)
print(model.predict([10.0]))
```
